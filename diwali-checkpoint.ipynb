# # Exploratory Data Analysis

# Gender

```python
ax = sns.countplot(x='Gender', data=df, hue='Gender', palette='Set2', legend=False)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sales_gen = df.groupby('Gender', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(data=sales_gen, x='Gender', y='Amount', hue='Gender', palette='Set2', legend=False)
plt.show()
```

# Age

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
    plt.show()
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group', y='Amount', data=sales_age, palette='Blues')
plt.show()
```

# State

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'] \
                .sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize': (15, 5)})
custom_colors = sns.color_palette("Set2", n_colors=len(sales_state))
ax = sns.barplot(data=sales_state,
                 x='State',
                 y='Orders',
                 hue='State',
                 palette=custom_colors,
                 legend=False)
for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation=5)
plt.tight_layout()
plt.show()
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'] \
                .sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize': (15, 5)})
custom_colors = sns.color_palette("Set2", n_colors=len(sales_state))
ax = sns.barplot(data=sales_state,
                 x='State',
                 y='Amount',
                 hue='State',          
                 palette=custom_colors,
                 legend=False)         
for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation=5)
plt.tight_layout()
plt.show()
```

# Marital Status

```python
sns.set(rc={'figure.figsize': (7, 5)})
ax = sns.countplot(data=df, x='Marital_Status', hue='Marital_Status',
                   palette='Set2', legend=False)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
plt.show()
```

```python
custom_colors = sns.color_palette("husl", df['Occupation'].nunique())
sns.set(rc={'figure.figsize': (20, 5)})
ax = sns.countplot(data=df, x='Occupation', hue='Occupation', palette=custom_colors, legend=False)
for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation=20)  
plt.tight_layout()
plt.show()
```

```python
custom_colors = sns.color_palette("husl", sales_state['Occupation'].nunique())

sns.set(rc={'figure.figsize': (20, 5)})

sns.barplot(data=sales_state, x='Occupation', y='Amount', hue='Occupation', palette=custom_colors, legend=False)

plt.xticks(rotation=15) 
plt.tight_layout()
plt.show()
```

# Product Category

```python
categories = df['Product_Category'].unique()
custom_colors = sns.color_palette("husl", len(categories))  # Or define your own color list
sns.set(rc={'figure.figsize': (20, 5)})
ax = sns.countplot(data=df, x='Product_Category', hue='Product_Category', palette=custom_colors, legend=False)
for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'] \
                .sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize': (20, 5)})
custom_colors = sns.color_palette("Set2", len(sales_state))
ax = sns.barplot(
    data=sales_state,
    x='Product_Category',
    y='Amount',
    hue='Product_Category',
    palette=custom_colors,
    legend=False
)
for bars in ax.containers:
    ax.bar_label(bars)

plt.xticks(rotation=5)
plt.tight_layout()
plt.show()
```

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'] \
                .sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize': (20, 5)})
custom_colors = sns.color_palette("husl", len(sales_state))
ax = sns.barplot(
    data=sales_state,
    x='Product_ID',
    y='Orders',
    hue='Product_ID',         
    palette=custom_colors,
    legend=False              
)
for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation=5)
plt.tight_layout()
plt.show()
```

```python
top_products = df.groupby('Product_ID')['Orders'] \
                 .sum().nlargest(10).sort_values(ascending=False)
custom_colors = sns.color_palette("Set2", n_colors=10)
fig1, ax1 = plt.subplots(figsize=(12, 7))
top_products.plot(kind='bar', color=custom_colors, ax=ax1)
ax1.set_title('Top 10 Products by Orders')
ax1.set_xlabel('Product ID')
ax1.set_ylabel('Total Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
